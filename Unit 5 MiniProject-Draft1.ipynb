## 3 - Create a Data Frame with no missing project theme description

```python
# The field mjtheme_namecode contains the code for the project theme. Note that each project may have more than one theme.
# Before counting projects under each theme we generate a dataframe with a column for the theme code, and a separate column for 
# the theme name. We also generate separate lines for every theme under each project.
# Note there seem to be duplicates in the list of themes.

mjtheme=WB_data['mjtheme_namecode']
mjtheme_df=pd.DataFrame()
for i in range(500):
    df_proj=pd.DataFrame(mjtheme[i])
    df_proj['rownum'] = i
    mjtheme_df=mjtheme_df.append(df_proj)

print("")
print("Code - Name combination after dropping duplicates but without substituting empty names with missing values")
print(mjtheme_df.info())
print(mjtheme_df[['code','name']].drop_duplicates().sort_values('code'))

print("")
print("Code - Name combination after substituting empty names with missing values and dropping duplicates")
mjtheme_df=mjtheme_df.replace(r'^\s*$', np.nan, regex=True).drop_duplicates()
print(mjtheme_df.info())
print(mjtheme_df[['code','name']].drop_duplicates().sort_values('code'))

```

```python
# Fill missing values (note empty theme names are not identified as missing value)

mjtheme_df=mjtheme_df.replace(r'^\s*$', np.nan, regex=True).drop_duplicates().sort_values(['code','name'])

mjtheme_df['MissName'] = mjtheme_df['name'].isnull().astype(int)
print(mjtheme_df.groupby('code')['MissName'].sum())
print(mjtheme_df.groupby('code')['name'].count())

maxfill = pd.DataFrame(Counter(mjtheme_df['code']).most_common()).iloc[-1,1]
print(maxfill)
#print(Counter(mjtheme_df['name'],nan=TRUE).most_common())

mjtheme_df = mjtheme_df.fillna(method='ffill',limit=maxfill)
print(mjtheme_df[['code','name']].drop_duplicates().sort_values('code'))
```
